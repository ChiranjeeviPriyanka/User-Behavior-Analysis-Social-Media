4. Select at least two algorithms and apply them to the dataset.

We will apply K-Means clustering algorithm and DBSCAN clustering algorithm

```python
#K-Means Clustering
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
#select features for clustering
#Drop User_ID since it's irrelevant for clustering and it is unique ID for users
features = df.drop(columns=['User_ID'])
#Standardizing the features
sc= StandardScaler()
features_scale= sc.fit_transform(features)
#Applying K-Means with Elbow Method
inertia = []
cluster_range = range(1, 11) #Testing 1 to 10 clusters
for i in cluster_range:
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(features_scale)
    inertia.append(kmeans.inertia_)#Store the within-cluster sum of squares
#Plotting the Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(cluster_range, inertia, marker='o', linestyle='-', color='b')
plt.title('Elbow Method for Optimal Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia (Sum of Squared Distances)')
plt.xticks(cluster_range)
plt.grid(True)
plt.show()
```

```python
#Considering 4 clusters from the elbow method
#evaluate K-Means Clustering
#Apply PCA for better visualization (reduce to 2 components for 2D plotting)
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
features_pca = pca.fit_transform(features_scale)
#Apply K-Means with 4 Clusters
kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(features_pca)
#Evaluate the model
from sklearn.metrics import silhouette_score,davies_bouldin_score
#compute inertia (sum of squared distances to the centriods)
inertia = kmeans.inertia_
silhouette_avg = silhouette_score(features_pca, df['Cluster'])
davies_bouldin = davies_bouldin_score(features_pca, df['Cluster'])
#Display metrics
print("Inertia (Sum of Squared Distances):", inertia)
print("Silhouette Score:", silhouette_avg)
print("Davies-Bouldin Index:", davies_bouldin)
```

```python
# Visualize the clusters with centroids
plt.figure(figsize=(8, 5))

# Plot each cluster with a different color
plt.scatter(features_pca[df['Cluster'] == 0, 0], features_pca[df['Cluster'] == 0, 1], s=50, c='red', label='Cluster 0')
plt.scatter(features_pca[df['Cluster'] == 1, 0], features_pca[df['Cluster'] == 1, 1], s=50, c='blue', label='Cluster 1')
plt.scatter(features_pca[df['Cluster'] == 2, 0], features_pca[df['Cluster'] == 2, 1], s=50, c='green', label='Cluster 2')
plt.scatter(features_pca[df['Cluster'] == 3, 0], features_pca[df['Cluster'] == 3, 1], s=50, c='orange', label='Cluster 3')


# Mark centroids
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, c='black', marker='X', label='Centroids')

# Set the plot title and axis labels
plt.title("K-Means Clustering Visualization with 4 Clusters", fontsize=16)
plt.xlabel("Principal Component 1", fontsize=14)
plt.ylabel("Principal Component 2", fontsize=14)

# Display the legend with the cluster labels and centroids
plt.legend(title="Cluster")

# Show the plot
plt.show()
```

```python
#User engagement through K-Means Clustering for Facebook platform
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
# Select only Facebook users (Platform == 1)
facebook_df = df[df['Platform'] == 0].copy()
# Drop 'User_ID' column as it's irrelevant for clustering
features = facebook_df.drop(['User_ID', 'Platform'], axis=1)
# Standardizing the features
sc = StandardScaler()
features_scaled = sc.fit_transform(features)
# Apply PCA for better visualization (reduce to 2 components for 2D plotting)
pca = PCA(n_components=2)
features_pca_fb = pca.fit_transform(features_scaled)
# Apply K-Means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
facebook_df['Cluster'] = kmeans.fit_predict(features_pca_fb)
# Now assign the clusters back to the original df
df.loc[df['Platform'] == 0, 'Cluster'] = facebook_df['Cluster']

# Visualize the clusters with centroids
plt.figure(figsize=(8, 5))
# Plot each cluster with a different color
for cluster_num in range(3):
    cluster_data = features_pca_fb[facebook_df['Cluster'] == cluster_num]
    plt.scatter(cluster_data[:, 0], cluster_data[:, 1], s=50, label=f'Cluster {cluster_num}')
# Mark centroids
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, c='black', marker='X', label='Centroids')
# Set the plot title and axis labels
plt.title("K-Means Clustering for Facebook Users", fontsize=16)
plt.xlabel("Principal Component 1", fontsize=14)
plt.ylabel("Principal Component 2", fontsize=14)
# Display the legend with the cluster labels and centroids
plt.legend(title="Cluster")
# Show the plot
plt.show()



```

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) algorithm

```python
#DBSCAN Clustering
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
# Drop 'User_ID' as it's irrelevant for clustering and it's just a unique ID for users
features = df.drop(columns=['User_ID'])
# Standardize the features (Scaling)
sc = StandardScaler()
features_scale_db = sc.fit_transform(features)

# Apply PCA for better visualization (reduce to 2 components for 2D plotting)
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
features_pca_db = pca.fit_transform(features_scale_db)


# Apply DBSCAN (you can tune the epsilon and min_samples parameters)
dbscan = DBSCAN(eps=0.6, min_samples=6)  # You can adjust these parameters
df['Cluster'] = dbscan.fit_predict(features_pca_db)


#Evaluate the model
from sklearn.metrics import silhouette_score,davies_bouldin_score
#Silhouette Score
silhouette_avg = silhouette_score(features_pca_db, df['Cluster'])
davies_bouldin = davies_bouldin_score(features_pca_db, df['Cluster'])
#Display metrics
print("Silhouette Score:", silhouette_avg)
print("Davies-Bouldin Index:", davies_bouldin)
```

```python
#Visualize the clusters with centroids (DBSCAN doesn't use centroids like K-Means)
#Plot the clusters with different colors
plt.figure(figsize=(8, 5))


#Noise points are marked as -1, so we color them separately (for example, black)
plt.scatter(features_pca_db[df['Cluster'] == -1, 0], features_pca_db[df['Cluster'] == -1, 1], s=50, c='black', label='Noise')

#Plot the rest of the clusters
plt.scatter(features_pca_db[df['Cluster'] == 0, 0], features_pca_db[df['Cluster'] == 0, 1], s=50, c='red', label='Cluster 0')
plt.scatter(features_pca_db[df['Cluster'] == 1, 0], features_pca_db[df['Cluster'] == 1, 1], s=50, c='blue', label='Cluster 1')
plt.scatter(features_pca_db[df['Cluster'] == 2, 0], features_pca_db[df['Cluster'] == 2, 1], s=50, c='green', label='Cluster 2')
plt.scatter(features_pca_db[df['Cluster'] == 3, 0], features_pca_db[df['Cluster'] == 3, 1], s=50, c='orange', label='Cluster 3')

# Set the plot title and axis labels
plt.title("DBSCAN Clustering Visualization", fontsize=16)
plt.xlabel("Principal Component 1", fontsize=14)  # X-axis label (PCA 1)
plt.ylabel("Principal Component 2", fontsize=14)  # Y-axis label (PCA 2)

# Display the legend with the cluster labels
plt.legend(title="Cluster")

# Show the plot
plt.show()
```
